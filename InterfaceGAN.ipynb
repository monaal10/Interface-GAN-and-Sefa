```python
import argparse
import torch
from torchvision import utils
from model import Generator
from tqdm import tqdm
import pickle5 as pickle
import pandas as pd
import os
```

```python
imgNo = 0
def generate(sample_z, g_ema,folder_num,img_number, device = 'cuda', mean_latent = None):

    with torch.no_grad():
        g_ema.eval()
        for i in range(1):
            
#             sample_z = torch.randn(1, 512, device=device)

            sample, _ = g_ema(
                [sample_z.to(device).float()], truncation=1.0 , truncation_latent=None
            )
            folder_dir = f"Output_Neutral/{folder_num}"
            CHECK_FOLDER = os.path.isdir(folder_dir)
            if not CHECK_FOLDER:
                os.makedirs(folder_dir)
            utils.save_image(
                sample[4],
                f"{folder_dir}/{str(img_number).zfill(6)}.png",
                nrow=1,
                normalize=True,
                range=(-1, 1),
            )
```

```python
g_ema = Generator(256, 512, 8, channel_multiplier=2).to('cuda')
ckpt = 'checkpoint/050000.pt'
checkpoint = torch.load(ckpt)
g_ema.load_state_dict(checkpoint["g_ema"])
```
